```python
#import pandas as pd

# URL of the CSV file

url = "https://raw.githubusercontent.com/C0D4IN/disciplina_ia_c1_2024/main/disciplina_ia_c1_2024/csvs/urls_phishing_checkpoint1.csv"

# Load the CSV file into a DataFrame
ds = pd.read_csv(url)

# Display the first few rows of the DataFrame to verify
ds.head(10)
```

```python
len(ds)
```

```python
df1 = ds.sample(n = 4000, random_state = 85)
len(df1)
```

```python
# Característica da base
ds.info()
```

```python
# @title Subset

# CLASS_LABEL: Rótulo da classe, onde
#              0 indica uma URL legítima e
#              1 indica uma URL de phishing.

# Dividir o DataFrame em dois subconjuntos
df1 = ds[ds['depth_url'] == 1] # phishing
df0 = ds[ds['depth_url'] == 0] # legítima
df1.sample(5, random_state = 85)
```

```python
# @title Média

# Média de todas as observações da base
mean_value = ds['depth_url'].mean()
print(f"A média da coluna 'depth_url' é: {mean_value} (Todas URLs)")

# Média das URLs - Phishing
mean_value1 = df1['depth_url'].mean()
print(f"A média da coluna 'depth_url' é: {mean_value1} (URLs Phishing)")

# Média das URLs - Legítima
mean_value0 = df0['depth_url'].mean()
print(f"A média da coluna 'depth_url' é: {mean_value0} (URLs Legítima)")
```

```python
# @title Mediana

# Mediana de todas as observações da base
median_value = ds['depth_url'].median()
print(f"A mediana da coluna 'depth_url' é: {median_value} (Todas URLs)")

# Mediana das URLs - Phishing
median_value1 = df1['depth_url'].median()
print(f"A mediana da coluna 'depth_url' é: {median_value1} (URLs Phishing)")

# Mediana das URLs - Legítima
median_value0 = df0['depth_url'].median()
print(f"A mediana da coluna 'depth_url' é: {median_value0} (URLs Legítima)")
```

```python
# @title Desvio Padrão

# Desvio Padrão de todas as observações da base
std_dev_value = ds['depth_url'].std()
print(f"O desvio padrão da coluna 'depth_url' é: {std_dev_value.round(2)} (Todas URLs)")

# Desvio Padrão das URLs - Phishing
std_dev_value1 = df1['depth_url'].std()
print(f"O desvio padrão da coluna 'depth_url' é: {std_dev_value1.round(2)} (URLs Phishing)")

# Desvio Padrão das URLs - Legítima
std_dev_value0 = df0['depth_url'].std()
print(f"O desvio padrão da coluna 'depth_url' é: {std_dev_value0.round(2)} (URLs Legítima)")
```
